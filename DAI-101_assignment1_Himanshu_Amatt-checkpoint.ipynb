### Statistical Tests

### Chi-Square Test

```python
# Create a contingency table for two categorical variables: isAdult and release_year
contingency_table = pd.crosstab(df['isAdult'], df['release_year'])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies: \n", expected)
```

### Conclusion
Since the p-value is higher than the typical threshold of 0.05, we fail to reject the null hypothesis. So we can that the data does not provide strong evidence that involvement in extracurricular activities is related to motivation level.

### T-Test

```python
# Create a new categorical variable based on runtime
df['runtime_category'] = df['runtimeMinutes'].apply(lambda x: 'Short' if x < 90 else 'Long')

# Separate data into two groups: Short and Long movies
group_short = df[df['runtime_category'] == 'Short']['averageRating'].dropna()
group_long = df[df['runtime_category'] == 'Long']['averageRating'].dropna()

# Perform the T-Test if both groups are valid
t_stat, p_val = ttest_ind(group_short, group_long)
print(f"T-Statistic: {t_stat}")
print(f"P-value: {p_val}")
```

### Conclusion
The negative value of the T-Statistic indicates that the average rating for short movies is significantly lower than that for long movies. The larger the absolute value of the T-Statistic, the more significant the difference between the two groups.The P-value of 8.3705e -16 is extremely low (much less than 0.05). This suggests that there is a statistically significant difference in the average ratings between short and long movies.

### Anova Test

```python
# Group the data by release_year
groups = [group['averageRating'].dropna() for name, group in df.groupby('release_year')]

f_stat, p_val = f_oneway(*groups)

print(f"F-Statistic: {f_stat}")
print(f"P-value: {p_val}")
```

### Conclusion 

F-Statistic: The F-statistic of 3.89 suggests that there is variability in the average ratings between the different release_year groups. A p-value of 3.34e-19 is significantly lower than the common alpha level of 0.05. This means that you reject the null hypothesis (which states that there is no difference in means among the groups) and conclude that there are significant differences in average ratings across different release years.
